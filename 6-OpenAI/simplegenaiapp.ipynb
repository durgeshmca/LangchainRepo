## Simple Gen AI App using langchain

```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
# Data ingestion from the website scrap the data
from langchain_community.document_loaders import WebBaseLoader
loader = WebBaseLoader("https://docs.smith.langchain.com/tutorials/Administrators/manage_spend")
loader
```

```python
docs = loader.load()
```

```python
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100)
documents = text_splitter.split_documents(docs)
documents
```

```python
# convert to vectors
from langchain_openai import  OpenAIEmbeddings
embeddings = OpenAIEmbeddings()
```

```python
from langchain_community.vectorstores import FAISS
```

```python
vectore_db = FAISS.from_documents(documents,embeddings)
```

```python
# Query from the vectore store db
query = "What are the usage limits of langsmith?"
result = vectore_db.similarity_search(query)
result[0].page_content
```
